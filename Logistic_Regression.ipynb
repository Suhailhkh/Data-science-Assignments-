```python
from sklearn.preprocessing import LabelEncoder
le_encoder=LabelEncoder()
bankclient_info['job']=le_encoder.fit_transform(bankclient_info['job'])
bankclient_info['marital']=le_encoder.fit_transform(bankclient_info['marital'])
bankclient_info['education']=le_encoder.fit_transform(bankclient_info['education'])
bankclient_info['default']=le_encoder.fit_transform(bankclient_info['default'])
bankclient_info['housing']=le_encoder.fit_transform(bankclient_info['housing'])
bankclient_info['loan']=le_encoder.fit_transform(bankclient_info['loan'])
bankclient_info['contact']=le_encoder.fit_transform(bankclient_info['contact'])
bankclient_info['month']=le_encoder.fit_transform(bankclient_info['month'])
bankclient_info['poutcome']=le_encoder.fit_transform(bankclient_info['poutcome'])
bankclient_info['y']=le_encoder.fit_transform(bankclient_info['y'])
bankclient_info
```

### Model Building

```python
x=bankclient_info.drop('y',axis=1)
y=bankclient_info[['y']]
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=12)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

### Model Training


```python
from sklearn.linear_model import LogisticRegression
logistic_model=LogisticRegression() #Model Initialisation/Estimator
logistic_model.fit(x_train,y_train)
```

### Model Testing | Model Evaluation

```python
y_train_pred=logistic_model.predict(x_train)
y_test_pred=logistic_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print('Accuracy Score :',accuracy_score(y_train,y_train_pred))
print('---------------------------------------------')
print('Confusion Matrix :\n',confusion_matrix(y_train,y_train_pred))
print('---------------------------------------------')
print(classification_report(y_train,y_train_pred))
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
fpr,tpr,thresholds= roc_curve(y,logistic_model.predict_proba (x)[:,1])

auc= roc_auc_score(y_train,y_train_pred)
print('Area Under the Curve[AUC]:',auc)
```

```python
import matplotlib.pyplot as plt
plt.plot(fpr,tpr,color='blue',label='ogistic_model(area=%0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False positive Rate or [1- True Negative Rate]')
plt.ylabel('True Positive Rate')
```
