```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
from sklearn import metrics
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
fd_info=pd.read_csv("Fraud_check.csv")
fd_info
```

```python
fd_info.describe()
```

```python
fd_info.dtypes
```

```python
le_encoder=LabelEncoder()
fd_info['Undergrad']=le_encoder.fit_transform(fd_info['Undergrad'])
fd_info['Urban']=le_encoder.fit_transform(fd_info['Urban'])
fd_info
```

```python
fd_info=pd.get_dummies(fd_info,columns=['Marital.Status'],drop_first=True)
fd_info
```

```python
fd_info=fd_info.assign(taxgrp=pd.cut(fd_info['Taxable.Income'],bins=[0,30000,100000],labels=['Risky','Good']))
fd_info
```

```python
fd_info=pd.get_dummies(fd_info,columns=['taxgrp'],drop_first=True)
fd_info
```

```python
sns.pairplot(data=fd_info,hue='taxgrp_Good')
plt.show()
```

Model Building

```python
x=fd_info.drop('taxgrp_Good',axis=1)
y=fd_info[['taxgrp_Good']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=12)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier()
dt_model.fit(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
grid_search_cv=GridSearchCV(estimator=dt_model,param_grid={'criterion':['entropy','gini'],'max_depth':[1,2,3,4,5,]},cv=5)
grid_search_cv.fit(x,y)
print(grid_search_cv.best_params_)
```

```python
#predictiom
y_pred=dt_model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(data=cm,linewidth=.5,annot=True,square=True,cmap='Blues')
plt.show()
```

```python
from sklearn import metrics
accuracy_score=metrics.accuracy_score(y_test,y_pred)
accuracy_score
```

Plot the tree

```python
from sklearn import tree
plt.figure(figsize=(8,13))
tree.plot_tree(dt_model,rounded=True,filled=True,max_depth=1)
plt.show()
```
