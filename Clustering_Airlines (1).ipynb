```python
import pandas as pd
import numpy as np
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score,calinski_harabasz_score
import warnings 
warnings.filterwarnings("ignore")
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return (x)
```

```python
airlines_norm =norm_func(airlines_info.iloc[:,1:])                 
airlines_norm
```

```python
dendrogram=sch.dendrogram(sch.linkage(airlines_norm,method='complete'))
```

```python
hc=AgglomerativeClustering(n_clusters=4,affinity='euclidean')
hc
```

```python
y_hc=hc.fit_predict(airlines_norm)
y_hc
```

```python
clusters=pd.DataFrame(y_hc,columns=['clusters'])
clusters
```

```python
airlines_info['clusters']=clusters
airlines_info
```

```python
plt.scatter(airlines_info['clusters'],airlines_info['Balance'],c=hc.labels_)
plt.show()
```

```python
airlines_info1=airlines_info.drop('clusters',axis=1)
airlines_info1
```

```python
nc=range(1,11)
inertia=[]
for i in nc:
    cluster1=KMeans(n_clusters=i)
    label=cluster1.fit_predict(airlines_norm)
    inertia.append(cluster1.inertia_   )
plt.scatter(nc,inertia)
plt.plot(nc,inertia)
inertia
plt.show()
```

```python
nc=range(2,11)
sil=[]
for i in nc:
    cluster2=KMeans(n_clusters=i)
    label=cluster2.fit_predict(airlines_norm)
    sil.append(silhouette_score(airlines_norm,label))
plt.scatter(nc,sil)
plt.plot(nc,sil)
sil
plt.show()
```

```python
max(sil)
```

```python
nc=range(2,10)
cal=[]
for i in nc:
    cluster3=KMeans(n_clusters=i)
    label=cluster3.fit_predict(airlines_norm)
    cal.append(calinski_harabasz_score(airlines_norm,label))
plt.scatter(nc,cal)
plt.plot(nc,cal)
cal
plt.show()
```

```python
kmean_cluster=KMeans(n_clusters=2)
cluster=kmean_cluster.fit_predict(airlines_norm)
cluster
```

```python
airlines_info1['cluster']=cluster
airlines_info1
```
